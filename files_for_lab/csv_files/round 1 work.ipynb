```python
# Read the three files into python as dataframes
import pandas as pd
file1 = pd.read_csv('file1.csv')
file2 = pd.read_csv('file2.csv')
file3 = pd.read_csv('file3.csv')
```

```python
# Show the DataFrame's shape.
print("file1", file1.shape)
print("file2", file2.shape)
print("file3", file3.shape)
```

```python
# Standardize header names.
# (first compare header names)
print(file1.columns)
print(file2.columns)
print(file3.columns)
```

```python
#(change all to lowercase)
cols = []
for column in file1.columns:
    cols.append(column.lower())
file1.columns = cols

cols2 = []
for column in file2.columns:
    cols2.append(column.lower())
file2.columns = cols2

cols3 = []
for column in file3.columns:
    cols3.append(column.lower())
file3.columns = cols3

#(rename columns)
file1 = file1.rename(columns={'st':'state'})
file2 = file2.rename(columns={'st':'state'})
```

```python
#Rearrange the columns in the dataframe as needed
file1 = file1[['customer', 'gender', 'state', 'education', 'income', 'customer lifetime value', 'policy type', 'vehicle class', 'monthly premium auto', 'number of open complaints', 'total claim amount']]
file2 = file2[['customer', 'gender', 'state', 'education', 'income', 'customer lifetime value', 'policy type', 'vehicle class', 'monthly premium auto', 'number of open complaints', 'total claim amount']]
file3 = file3[['customer', 'gender', 'state', 'education', 'income', 'customer lifetime value', 'policy type', 'vehicle class', 'monthly premium auto', 'number of open complaints', 'total claim amount']]
```

```python
#Concatenate the three dataframes
data = pd.concat([file1,file2,file3], axis=0)
data
```

```python
# Which columns are numerical?
# Which columns are categorical?
data.dtypes
```

```python
# Understand the meaning of all columns
# Perform the data cleaning operations mentioned so far in class
# Delete the column education and the number of open complaints from the dataframe.
data = data.drop(['education', 'number of open complaints'], axis=1)
data
```

```python
# Correct the values in the column customer lifetime value. 
# They are given as a percent, so divide them by 100 and change dtype to numerical type.
data['customer lifetime value'] = pd.to_numeric(data['customer lifetime value'], errors='coerce')

data['customer lifetime value'] = data['customer lifetime value']/100
data
```

```python
# Check for duplicate rows in the data and remove if any.
data = data.drop_duplicates()
data
```

```python
# Filter out the data for customers who have an income of 0 or less.
filtered = data[data['income']>0]
filtered
```
